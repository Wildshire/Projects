# Auxiliar Functions

```python
'''
Read the audio file properly, there are problems if we do not cast to double
wavfile_name: the path to the wavfile
'''
def read_file_PROPERLY(wavfile_name):
    samplerate_music, data_music = wavfile.read(wavfile_name) #Read the file
    data_music=data_music.astype(np.double)/2**15
    return samplerate_music,data_music
```

```python
'''
Return the masked data with a mask omega 
'''
def Phi(f,omega):
    return f*(1-omega)
```

# Direct problem generator
We are going to create the image by first creating the binary mask, apply it to the image and then sum some white noise.

```python
observed_audios=[]
omegas=[]
#stds = [0.42,0.25, 0.16, 0.08, 0.04] #White noises
#probabilities=np.arange(0,1,0.1)
stds=[0.04]
probabilities=[0.3]
for p in probabilities:
    omega=np.random.binomial(1,p,(len(data_music)))
    omegas.append(omega)
    masked=Phi(data_music,omega)
    for std in stds:
        noise = np.random.normal(0, 1, size=data_music.shape)
        obs_aud = masked + std*noise 
    observed_audios.append(obs_aud)
```

```python
ipd.Audio(observed_audios[0], rate=samplerate_music)
```

```python
plt.figure(figsize=(10,10))
plt.plot(observed_audios[0])
```
